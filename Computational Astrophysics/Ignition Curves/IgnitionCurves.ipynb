```python
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def q_H(T, X=0.7, Z=0.02, q_min=1.e3):
    'Return the density for the given temperature for H burning'
    T9 = T/1.e9
    # q_pp = ÏA(T) #
    A = 2.4e4*X**2*T9**(-2./3.)*np.exp(-3.38/T9**(1./3.))
    # q_cno = ÏB(T) #
    B = 4.4e25*X*Z*T9**(-2./3.)*np.exp(-15.228/T9**(1./3.))
    return q_min/(A+B)

def q_He(T, Y=1.0, q_min=1.e3):
    'Return the density for a given temperature for He burning'
    T8 = T/1.e8
    A = 5.09e11*Y**3*T8**(-3)*np.exp(-44.027/T8)
    return np.sqrt(q_min/A)

T = np.logspace(6.7,7.5,100)
rho_H = q_H(T)
T_He = np.logspace(7.9,8.3,100)
rho_He = q_He(T_He)
T_sun = 1.5e7
rho_sun = 160
plt.title('Ignition Curves for H and He')
plt.xlabel('Log(T [K])')
plt.ylabel('Log(Ï [g cm$^{-3}$])')
plt.loglog(T, rho_H, label='H Burning')
plt.loglog(T_He, rho_He, label='He Burning')
plt.scatter(T_sun, rho_sun, label='Solar Core')
plt.legend(frameon=False)
plt.show()
```
